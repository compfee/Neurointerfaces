3.	Собрать установку для измерения данных ФПГ, передать данные на ПК с помощью pySerial, вычислить основные параметры вариационной пульсометрии. Замеры провести в трех состояниях (перед нагрузкой, после 5 минут бега, после 15 минут бега)


```python
import pandas as pd
```

```python
df = pd.read_table("FPG1.dat", sep='	', names=["time", "value"], header=None)
df.reset_index(inplace = True)
df.pop("time")
df.head()
```

```python
from matplotlib import pyplot as plt
plt.plot(df['value'][:1000])
plt.title('FPG')
```

```python
from scipy.signal import find_peaks

rPeaks, _ = find_peaks(df["value"], height = 1.5, distance = 150)
rPeaks = rPeaks.tolist()
plt.figure(figsize=(23, 9))
plt.hlines(0.5, 0, len(df["value"]))
plt.plot(df["value"])
plt.plot(rPeaks, df["value"][rPeaks], "x")
plt.legend(loc='best')
plt.show()
```

```python
rIntervalsFPG = []
for i in range(len(rPeaks) - 1):
    rIntervalsFPG.append((rPeaks[i + 1] - rPeaks[i]) / 300 * 1000)
rIntervalsFPG.append(pd.Series(rIntervalsFPG).mean())
```

```python
plt.figure(figsize=(18, 6))
plt.plot(rIntervalsFPG, marker='o')
plt.show()
```

```python
seriesIntervalsECG = pd.Series(rIntervalsFPG)

MECG = seriesIntervalsECG.mean() / 1000
print("Математическое ожидание        = " + str(MECG) + " с")

SECG = seriesIntervalsECG.std(ddof = 0) / 1000
print("Среднеквадратичное отклонение  = " + str(SECG) + " с")

CVECG = SECG / MECG * 100
print("Коэффициент вариации           = " + str(CVECG))
```

```python
plt.hist(x = rIntervalsFPG, label='FPG')
plt.legend(loc='upper right')
plt.show()
```
